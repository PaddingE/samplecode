```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## 막대 그래프

x = np.arange(3)
y_value = [100,400,600]
y_std = [20,34,22]
plt.bar(x,y_value, color = ['m','r','b'], alpha = 0.6,
        edgecolor = 'r',
        width = 0.6, yerr = y_std, capsize = 3, ecolor = 'b')

plt.title('Title')
plt.xlabel('Year')
plt.ylabel('Value')
plt.xticks(x, labels=['2019', '2020', '2021'])

plt.show()
```

```python
y = np.arange(3)
x_value = [100,400,600]
y_std = [20,34,22]
plt.barh(y,x_value, color = ['m','r','b'], alpha = 0.6
        )

plt.title('Title')
plt.ylabel('Year')
plt.xlabel('Value')
plt.yticks(y, labels=['2019', '2020', '2021'])

plt.show()
```

```python
x = np.random.normal(10,1,100)
y_cor = x + np.random.randn(100)
y_size = np.random.normal(100,100,100)
y = np.random.randn(100)

print(np.corrcoef(x,y_cor))

plt.scatter(x,y_cor, s = y_size, alpha= 0.5)

plt.show()
```

```python
##
plt.scatter([1,2],[2,2], s = 100, c = ['r', 'b'],
            alpha= 0.6)
plt.text(1,2.02,'color: red', size = 14)
plt.text(2,2.02,'color: blue', size = 14)
plt.xlim(0,3)
plt.show()
```

```python
## 히스토그램

x = np.random.randn(1000)
y = np.random.uniform(-3,3,1000)

plt.hist(x, bins = 30, label = 'Normal dist', alpha = 0.5)
plt.hist(y, bins = 30, label= 'Uniform dist', alpha = 0.5)
plt.show()
```

```python
## boxplot

a = np.random.normal(1,2,100)
b = np.random.normal(3,2,100)
c = np.random.normal(5,2,100)

plt.boxplot([a,b,c], whis=2.5, notch = True, vert = False,
            showmeans= True)

plt.show()
```

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
titanic.dtypes
```

```python
fig, ax = plt.subplots(1,2)

#ax[0].scatter(titanic['age'],titanic['fare'])
#ax[1].scatter(titanic['age'],titanic['fare'])

sns.regplot(x = 'age', y = 'fare', data = titanic,
            marker = '+',
            ax = ax[0], fit_reg = False)
sns.regplot(x = 'age', y = 'fare', data = titanic,
            ax = ax[1], fit_reg = True)

plt.show()
```

```python
## distplot
sns.set_style('darkgrid')

fig, ax = plt.subplots(1,3, figsize = (12,5))

sns.distplot(titanic['fare'], ax = ax[0])
sns.distplot(titanic['fare'], hist = False, ax = ax[1])
sns.distplot(titanic['fare'], kde = False, ax = ax[2])
ax[0].set_title('dist + kde')
ax[1].set_title('kde')
ax[2].set_title('dist')
plt.show()
```

```python
## heatmap

titanic.head()



table = titanic.pivot_table(index = ['sex'], columns= ['class'], aggfunc = 'size')
sns.heatmap(table, annot= True)

plt.show()
```

```python
## strip, swarm

fig, ax = plt.subplots(1,2,figsize = (12,6))

sns.stripplot(x = 'class', y = 'age', data = titanic,
              ax = ax[0])
sns.swarmplot(x = 'class', y = 'age', data = titanic,
              ax = ax[1])

plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize = (12,6))

sns.barplot(x = 'sex', y = 'survived',data = titanic,
            ax = ax[0])
sns.barplot(x = 'sex', y = 'survived',data = titanic,
            hue = 'embarked',
            ax = ax[1])

plt.show()
```

```python
## 

fig, ax = plt.subplots(1,3,figsize = (12,6))

sns.boxplot(x = 'alive', y = 'age',data = titanic,
            ax = ax[0])
sns.boxplot(x = 'alive', y = 'age',data = titanic,
            hue= 'sex',
            ax = ax[1])
sns.violinplot(x = 'alive', y = 'age',data = titanic,
            hue= 'sex',
            ax = ax[2])
```

```python
df = pd.read_excel('./서울지역 대학교 위치.xlsx',
                   index_col= 0, header= 0)

seoul_map = folium.Map(location = [37.55, 126.98],
                       zoom_start = 12)

#df.head()

for name, lat, lng in zip(df.index, df['위도'], df['경도']):
    folium.CircleMarker([lat, lng], radius= 8,
                        color = 'brown',
                        fill = True,
                        fill_opacity = 0.5,
                        popup = name).add_to(seoul_map)

seoul_map
```
