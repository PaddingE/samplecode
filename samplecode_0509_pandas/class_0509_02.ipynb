```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib.font_manager as fm
from matplotlib import rc
import matplotlib as mpl
```

```python
## 한글설정
path = 'C:/Windows/Fonts/H2GTRE.TTF'
font_name = fm.FontProperties(fname= path).get_name()
rc('font', family = font_name)
#print(font_name)

##음수값 설정
mpl.rcParams['axes.unicode_minus'] = False
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.figure(figsize=(4,3))
plt.plot(x,y)
plt.show()
```

```python
data_dict = {
    'data_x':[1,2,3,4],
    'data_y':[1,4,9,16]
}

plt.plot('data_x', 'data_y', data = data_dict)

my_fontdict= {'family':'monospace',
                      'color':'r',
                      'fontsize':20,
                      'weight':'bold'}

plt.xlabel('X', labelpad = 10, fontdict= my_fontdict)

plt.ylabel('X^2')
plt.title('X vs X^2')
plt.show()
```

```python
plt.plot('data_x', 'data_y', data = data_dict,
         label = 'Price ($)')
plt.xlabel('X', loc = 'right')
plt.ylabel('Y', loc = 'top')

plt.legend(loc = 'upper center')

plt.show()
```

```python

plt.figure(figsize = (4,3))
plt.plot(np.arange(1,100), np.arange(1,100)**2, label = 'x^2')
plt.plot(np.arange(1,100), np.arange(1,100)**2.3, label = 'x^2.3')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
plt.legend(frameon = True,
           edgecolor = 'r')
plt.show()
```

```python
#line type

plt.figure(figsize = (4,3))
plt.plot(np.arange(1,100), np.arange(1,100)**2, label = 'Dashed',
         linestyle = '--')
plt.plot(np.arange(1,100), np.arange(1,100)**2.3, label = 'Dotted',
         linestyle = ':')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
plt.legend(frameon = True,
           edgecolor = 'r')
plt.show()
```

```python
plt.figure(figsize= (6,4))
plt.plot([1,2,3],[4,4,4],linestyle = '-', marker = '+',
         color = 'r', label = 'Solid', linewidth = 2)
plt.plot([1,2,3],[3,3,3],linestyle = '--', marker = '.',
         color = 'b', label = 'Dashed')
plt.plot([1,2,3],[2,2,2],linestyle = ':', marker = 'o',
         color = 'c', label = 'Dotted')
plt.plot([1,2,3],[1,1,1],linestyle = '-.', marker = '^',
         color = 'm', label = 'Dash-dotted')
plt.ylim(0,5)
plt.legend(ncol = 2, loc = 'upper center')

plt.show()
```

```python
plt.figure(figsize= (6,4))
plt.plot([1,2,3],[4,4,4], color = '#b2c5d9')
plt.plot([1,2,3],[3,3,3], color = 'violet')
plt.plot([1,2,3],[2,2,2], color = 'dodgerblue')
plt.plot([1,2,3],[1,1,1])
plt.ylim(0,5)
plt.show()
```

```python
t = np.array([1,2,3,4])

plt.plot(t,t,'ro')
plt.plot(t,t**2,'g^--')
plt.plot(t,t**3,'bx-.')

plt.show()
```

```python
x = np.arange(0,2,0.2)

plt.plot(x,x,'bo')
plt.plot(x,x**2,'r^--', markersize = 2, linewidth = 2)

plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks([0,1,2], labels= ['Jan', 'Feb', 'Mar'])
plt.yticks(np.arange(0, 4), labels= ['100$', '200$', '300$', '400$'])

plt.tick_params(axis = 'x', direction = 'in',
                length = 3, pad = 6, labelsize = 10,
                labelcolor = 'm', labelrotation = 45)

plt.show()
```
